# Build a recommender system by using cosine simillarties score.

```python
import pandas as pd
import numpy as np
```

```python
books = pd.read_csv('book.csv' , encoding= 'Latin1') 
books
```

```python
#number of unique users in the dataset
len(books['User.ID'].unique())
```

```python
len(books['Book.Title'].unique()) #number of unique books in the dataset
```

```python
books_df = books.pivot_table(index='User_ID',
                                 columns='Book_Title',
                                 values='Book_Rating').reset_index(drop=True)
```

```python
books_df 
```

```python
books_df.index = books['User_ID'].unique()
```

```python
books_df
```

```python
books_df.fillna(0, inplace=True)
```

```python
books_df
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
# Calculating Cosine Similarity between Users on array data
user_sim = 1 - pairwise_distances( books_df.values,metric='cosine')
```

```python
user_sim
```

```python
#Store the results in a dataframe
user_sim_df = pd.DataFrame(user_sim)
```

```python
user_sim_df
```

```python
#Set the index and column names to user ids 
user_sim_df.index = books['User_ID'].unique()
user_sim_df.columns = books['User_ID'].unique()
```

```python
user_sim_df
```

```python
np.fill_diagonal(user_sim, 0)
user_sim_df
```

```python
#Most Similar Users
user_sim_df.idxmax(axis=1)
```
